```python
len(df.loc['2016':'2018'])
```

```python
train = df['alch'].loc[:'2016']
test=df['alch'].loc['2017':]
```

```python
%%time
m1=SARIMAX(train,order=(0,1,1),seasonal_order=(0,1,1,12)).fit()
m1.aic
```

```python
%%time
m2=SARIMAX(train,order=(1,1,1),seasonal_order=(0,1,1,12)).fit()
m2.aic
```

```python
%%time
m3=SARIMAX(train,order=(5,1,5),seasonal_order=(5,1,5,12)).fit()
m3.aic
```

```python
%%time
m4=SARIMAX(train,order=(11,1,5),seasonal_order=(5,1,7,12)).fit()
m4.aic
```

```python
%%time
m5=SARIMAX(train,order=(30,1,5)).fit()
m5.aic
```

```python
%%time
m6=SARIMAX(train,order=(8,1,4),seasonal_order=(0,1,7,12)).fit()
m6.aic
```

```python
%%time
m7=SARIMAX(train,order=(6,1,5),seasonal_order=(0,1,6,12)).fit()
m7.aic
```

```python
start=len(train)

end=start+50
```

```python
pred1=m1.predict(start,end).rename('pred1 ~ order=(0,1,1),seasonal_order=(0,1,1,12)')
pred2=m2.predict(start,end).rename('pred2 ~ order=(1,1,1),seasonal_order=(0,1,1,12)')
pred3=m3.predict(start,end).rename('pred3 ~ order=(5,1,5),seasonal_order=(5,1,5,12)')
pred4=m4.predict(start,end).rename('pred4 ~ order=(11,1,5),seasonal_order=(5,1,7,12)')
pred5=m5.predict(start,end).rename('pred5 ~ order=(30,1,5)')
pred6=m6.predict(start,end).rename('pred6 ~ order=(8,1,4),seasonal_order=(0,1,7,12)')
pred7=m7.predict(start,end).rename('pred7 ~ order=(6,1,5),seasonal_order=(0,1,6,12)')
```

```python
pred1.plot(legend=True,color='darkorange')
pred2.plot(legend=True,color='yellow')
pred3.plot(legend=True,color='blue')
pred4.plot(legend=True,color='darkorchid')
pred5.plot(legend=True,color='#df34eb')
pred6.plot(legend=True,color='#0aff02')
pred7.plot(legend=True,color='#34eba1')


test.plot(legend=True,color='red',label='test')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def mse(pred):
    print(f"MODEL: {pred.name}")
    print("MSE: ",mean_squared_error(test,pred))
    print("rMSE: ",mean_squared_error(test,pred)**0.5)
    print("Mean of Test: ",test.mean())
```

```python
mse(pred1)
```

```python
mse(pred2)
```

```python
mse(pred3)
```

```python
mse(pred4)
```

```python
mse(pred5)
```

```python
mse(pred6)
```

```python
mse(pred7)
```

```python
df['alch'].plot(legend=True, label='Actual')
m5.forecast(64).plot(legend=True, label='m5')
m6.forecast(64).plot(legend=True,label='m6')
m7.forecast(64).plot(legend=True,label='m7')
```

# <font color = 'slateblue'> Due to model 6 ``m6`` carrying the least aic and yielding the best rMSE (755.264), we are to use this model on the full dataset `df['alch']`

```python
%%time
final6=SARIMAX(df['alch'],order=(8,1,4),seasonal_order=(0,1,7,12)).fit()
final6.aic
```

```python
ax=df['alch'].plot(legend=True,label='USA Alcohol Sales by Millions of Dollars',color='darkorchid')
ax=final6.forecast(60).rename('5 year Forecast').plot(legend=True,color='limegreen')
ax.set_ylabel("Millions of Dollars")
ax.set_xlabel("Year")
ax.set_title("5 Year Forecasted Alcohol Sales in USA")
```

```python
ax=df['alch'].plot(legend=True,label='USA Alcohol Sales by Millions of Dollars',color='darkorchid')
ax=final6.forecast(120).rename('10 year Forecast').plot(legend=True,color='#f10c45')
ax.set_ylabel("Millions of Dollars")
ax.set_xlabel("Year")
ax.set_title("10 Year Forecasted Alcohol Sales in USA")
```

```python
ax=df['alch'].plot(legend=True,label='USA Alcohol Sales by Millions of Dollars',color='darkorchid')
ax=final6.forecast(300).rename('25 year Forecast').plot(legend=True,color='#ff0490')
ax.set_ylabel("Millions of Dollars")
ax.set_xlabel("Year")
ax.set_title("25 Year Forecasted Alcohol Sales in USA")
```
